```python
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
import matplotlib.pyplot as plt

def First_order(y, t, k, u):
    tau = 5.0
    dydt = (-y + k*u)/tau
    return dydt

t = np.linspace(0,10,11)
k = 2.0
u = np.zeros(len(t))
u[3:] =1.0
y0 = 0

ys = np.zeros(len(t))
ys[0] = y0

for i in range(len(t)-1):
    ts = [t[i], t[i+1]]
    y_ = odeint(First_order, y0, t,args=(k,u[i]))
    y0 = y_[1]
    ys[i+1] = y0
    
plt.plot(t,ys)
```
